```julia
using Plots
```

```julia
df_obj = [2327,2727,2795,3070,3124,2643,2677,3127,3135,3163,
    3156,3402,3512,3548,3601,3023,3043,3060,3135,3206];
```

```julia
x_axis = [i for i in 10:29];
```

```julia
sp_obj = [2262,2658,2717,2987,3035,2539,2566,3012,3014,3036,
3028,3257,3364,3399,3445,2845,2856,2867,2933,3003];
```

```julia
plot(x_axis, [df_obj, sp_obj], title="Objective Values for DF and SP", label = ["DF Obj" "SP Obj"], legend=:topleft)
xlabel!("Number of jobs")
ylabel!("Total distance")
```

```julia
savefig("df_obj and sp_obj.png")
```

```julia
optimality_gap = df_obj .- sp_obj;
```

```julia
plot(x_axis, optimality_gap, title="Objective Gap between DF and SP", label = ["Obj Gap"], legend=:topleft)
xlabel!("Number of jobs")
ylabel!("Objective gap")
```

```julia
savefig("objective gap DF SP.png")
```

```julia
plot(x_axis, optimality_gap ./ df_obj, title="Objective Gap as Proportion of DF Solution", label = "proportion", legend=:topleft)
xlabel!("Number of jobs")
ylabel!("Proportion difference")
```

```julia
savefig("objective gap proportion DF SP.png")
```
